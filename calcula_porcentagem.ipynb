```python
def calculate_percentage(values):
  for value in values:
    print(f"${value[0] * (value[1] / 100)}")

def add_percentage(values):
  for value in values:
    print(f"${value[0] + (value[0] * (value[1] / 100))}")

def reduce_percentage(values):
  for value in values:
    print(f"${value[0] - (value[0] * (value[1] / 100))}")
```

```python
reduce_percentage([(500, 10), (800, 25), (260, 38)])
```
